```python
def summary_annotations_by_title_df(title):
    
 
    uploaded = uploaded_df[uploaded_df.title==title]

    ann = annotated_df[annotated_df.title==title]


    uploaded_tokens = []
    uploaded_ners = []
    
    for i,row in uploaded.iterrows():

        for token, ner in zip(row['token'], row['ner_label']):

            uploaded_tokens.append(token.result)

            uploaded_ners.append(ner.result)

    ann_tokens = []
    ann_ners = []
    
    for i,row in ann.iterrows():

        for token, ner in zip(row['token'], row['ner_label']):

            ann_tokens.append(token.result)
            ann_ners.append(ner.result)

    summary = pd.DataFrame({'uploaded_token':uploaded_tokens, 'ann_token':ann_tokens,
                           'uploaded_ners':uploaded_ners, 'ann_ners':ann_ners})
    
    return summary
```

```python
# find different annotations between 2 dataset
diff_titles= []
missing_text = []
k=0

for i in range (0, 298):
    
    diff_dict = {}
    diff_dict[i] = []
    
    try:
        
        for j in range (0, len(summary_annotations_by_title_df(titles[i]))):
            
            if (summary_annotations_by_title_df(titles[i]).uploaded_ners[j]) != (summary_annotations_by_title_df(titles[i]).ann_ners[j]) :
                
                diff_dict[i].append(j)
                k = k+1

        diff_titles.append(diff_dict)
   
    except :
        
        missing_text.append(i)
```

```python
# different annotation list to dict
new_diff_dict = {}
for item in diff_titles:
    for key,values in item.items():
        new_diff_dict[key]=values

new_diff_dict
```

```python
import json
with open("new_diff_dict.txt", "w") as fp:
    json.dump(new_diff_dict,fp)
```

```python
# difference annotated text values to dict then df

difference_dict = {'title':[], 'text': [], 'token_num':[], 'uploaded_token':[], 'ann_token':[], 'uploaded_ners':[], 'ann_ners':[]  }

for key,values in new_diff_dict.items():
    
    
    
    for i in range (len(new_diff_dict[key])) :
        
        j = new_diff_dict[key][i]
        difference_dict['text'].append(uploaded_df.text.iloc[key])
        difference_dict['title'].append(key)
        difference_dict['token_num'].append(j)
        difference_dict['uploaded_token'].append(summary_annotations_by_title_df(titles[key]).loc[j][0])
        difference_dict['ann_token'].append(summary_annotations_by_title_df(titles[key]).loc[j][1])
        difference_dict['uploaded_ners'].append(summary_annotations_by_title_df(titles[key]).loc[j][2])
        difference_dict['ann_ners'].append(summary_annotations_by_title_df(titles[key]).loc[j][3])

# dict to df
difference_df = pd.DataFrame.from_dict(difference_dict)
difference_df.head()
```
